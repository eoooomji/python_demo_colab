## Data Scaling(데이터 스케일링)

- 피처(feature) = 컬럼(column)들마다 데이터의 값이 범위가 다 제각각이거나 단위가 다른 경우 데이터를 학습을 할 때 0으로 수렴하거나 무한으로 발산할 수 있다.

- 이런 경우 데이터 스케일링을 통해 모든 피처들의 데이터 분포나 범위를 동일하게 조정해 줄 수 있다.

- StandardScaler 
    - 평균 0, 표준편차 1
    - $z = \frac{x-\mu}{\sigma}$, $\mu$는 평균 $\sigma$는 표준편차
- MinMaxScaler
    - 최솟값 0, 최댓값 1
    - $z = \frac{x-x_{min}}{x_{max}- x_{min}}$


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

#### label encoding

- label encoding은 문자열 데이터에 고유한 라벨 번호를 매기는 것이다.

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sample(5)
```

```python
# 문자열 데이터
iris.species
```

```python
iris.species.map({'setosa' : 0, 'versicolor' : 1, 'virginica' : 2})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
ll = LabelEncoder()
```

```python
# 1차원으로 넣어도 된다.
# preprocessing은 fit_transfrom으로 한번에 할 수 있다.
ll.fit_transform(iris.species)

# ll.fit_tranform(iris['species'])
```

```python
# 2차원으로 넣으면 warning이 뜬다.
# sklearn이 체계적이지 않은 멶이 있다.
# ravel() 또는 column_or_1d를 사용하라고 뜬다.
ll.fit_transform(iris[['species']])
```

```python
from sklearn.utils import column_or_1d
```

```python
column_or_1d(iris['species'])
ll.fit_transform(column_or_1d(iris['species']))
```

```python
# inverse_transform() : 라벨(정수)를 다시 문자로 복원할 수 있다.
ll.inverse_transform([0,1 , 1, 1, 2, 2, 0])
```

### one hot encoding

- 문자열 데이터를 column으로 바꾸고, 해당하는 위치에만 1로 한다.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
# return sparse matrix 이다.
# 0이 많은 희소행렬은 메모리 효율성을 위해서 압축된 형식의 sparse matrix로 만든다.
ohe.fit_transform(iris[['species']])
```

```python
# toarray()는 numpy로 바꿔주어 값을 확인할 수 있다.
ohe.fit_transform(iris[['species']]).toarray()
```

```python
# 반대로 숫자를 문자로 바꿀 경우
ohe.inverse_transform([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
```
